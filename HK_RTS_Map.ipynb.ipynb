```python
from pykml import parser
import pandas as pd

# 读取KMZ文件
with open('./兩餸飯店.kml', 'rb') as f:
    kml_data = f.read()
```

```python


doc = parser.fromstring(kml_data)

# 获取Placemark元素
placemarks = doc.Document.Placemark

# 将地点信息存储到DataFrame中
data = []
for p in placemarks:
    name = str(p.name)
    coordinates = str(p.Point.coordinates)
    longitude, latitude, _ = coordinates.split(',')
    # long '\n          114.1995529' convert to float
    longitude = float(longitude.split('\n          ')[1])
    data.append([name, latitude, longitude])

df = pd.DataFrame(data, columns=['Name', 'Latitude', 'Longitude'])
df

```

```python
# astype float
df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)
```

```python
# leaf map 
import folium
from folium.plugins import MarkerCluster

# 创建地图
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=12)

#创建MarkerCluster对象
marker_cluster = MarkerCluster().add_to(m)

# 将地点信息添加到地图中
for i in range(len(df)):
    folium.Marker(
        location=[df.iloc[i]['Latitude'], df.iloc[i]['Longitude']],
        popup=df.iloc[i]['Name'],
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(marker_cluster)

# 保存地图
m.save('./map.html')
```
